```python
def datecreator(t1,t2):
  date_list=[]
  t = timedelta(days = 1)
  txt=""
  newdate=""
  url_in=""
  # array_page_link=[]
  url="https://www.sabah.com.tr/timeline/"
  dates = np.arange(t1, t2, t).astype(datetime)
  categories = ["gundem","ekonomi","yasam","dunya"]
  for i in categories:
    for j in dates:
      # print(i.strftime('%d-%m-%Y'))
      newdate=j.strftime('%Y/%m/%d')
      url_in="{}{}{}{}".format(url,newdate,"?c=",i)
      #print(url_in)
      date_list.append(url_in)
  
  return date_list
```

```python
def create_date_D1():
  date_list = []
  for year in range(2009,2021):
      for month in range(1,13):
        for day in range(1,31):
          #date = (str(year)+"/"+str(month)+"/"+str(day))
          date = "{}/{}/{}".format(year,month,day)
          date_list.append(date)
  return date_list
```

```python
def write_to_txt(data):
  with open("sabah1.txt", 'a') as file: 
    file.write(data+'\n')
    file.truncate()
```

```python
def get_data_from_url_D1(url):
  r = requests.get(url)   
  soup = BeautifulSoup(r.content, 'html5lib')
  title = soup.find("h1", attrs = {"class":"pageTitle"}).getText()
  content_array = soup.find("div", attrs = {"class":"newsBox"}).getText()
  date = soup.find("span", attrs = {"class":"textInfo"}).getText()
  date = date.split()
  date = date[2]
  content_array = content_array.split()
  content_string = ""
  for w in content_array:
    content_string = content_string+" "+w
  #w_data = url+";"+date+";"+title+";"+content_string
  w_data="{};{};{};{}".format(url,date,title,content_string)
  print(w_data)
  write_to_txt(w_data)
```

```python
def datecreator_D2(d1,d2):
  url_list = []
  t = timedelta(days = 1)
  url = "http://arsiv.sabah.com.tr"
  dates = np.arange(d1,d2,t).astype(datetime)
  categories = ["eko00.html","gnd00.html","siy00.html","dun00.html","spo00.html"]
  for category in categories:
    for date in dates:
      newdate=date.strftime('%Y/%m/%d')
      url_tmp = "{}/{}/{}".format(url,newdate,category)
      url_list.append(url_tmp)
  return url_list
```

```python
time1=datetime(2003,12,5)
time2=datetime(2007,4,5)
page_link_D2 = datecreator_D2(time1,time2)
for i in page_link_D2:
  print(i)
```

```python
news_links_D2 = []
def find_link_in_page_D2(link):
  r = requests.get(link)   
  soup = BeautifulSoup(r.content, 'html5lib')
  link2 = soup.findAll("a")
  categories1 = ["eko00","gnd00","siy00","dun00","spo00"]
  categories2 = ["eko","gnd","siy","dun","spo"]
  for l in link2:
    href = l.get('href')
    try:
      if href.startswith(tuple(categories2)) and  not(href.startswith(tuple(categories1))):
        url_temp = link[0:-11]
        url_temp = "{}/{}".format(url_temp,href)
        news_links_D2.append(url_temp)
        #print(url_temp)
    except:
      pass
```

```python
def get_data_from_url_D2(url):
  r = requests.get(url)   
  soup = BeautifulSoup(r.content, 'html5lib')
  title = soup.find("span", attrs = {"class":"detay_baslik"}).getText()
  title = title.splitlines()
  title = title[0].strip()
  content = soup.find("span", attrs = {"class":"detay_haber"}).getText()
  content_array = content.split()
  content_string = ""
  for w in content_array:
    content_string = content_string+" "+w
  date = url[26:36]
  date = date.split('/')
  date = date[2]+"."+date[1]+"."+date[0]
  w_data="{};{};{};{}".format(url,date,title,content_string)
  write_to_txt(w_data)
```
